```python
# assigning the path and list of age categories
# set the path of the folder where the images stored
flood_data_HL_dir = "/content/drive/MyDrive/Sen1Flood11_New_Dataset/train/1/"
# flood_data_HL_dir = '/Users/kiranbijoy/Sem_3/Main Project/ML algorithms/train/1/'



# set the list of categories available in the folder
flood_img_data = []

# iterate over each images, considering all the age groups
for img in os.listdir(flood_data_HL_dir):
  # read one by one image
    # print(img)
    try:
      img_array = keras.utils.load_img(os.path.join(flood_data_HL_dir, img), target_size=(32, 32))
    except:
      pass

    image = img_to_array(img_array)
    # # reshape data for the model

    image = image.reshape(( image.shape[0], image.shape[1], image.shape[2]))
    image = image.astype('float32')
    image /= 255

    # image = preprocess_input(image)
    flood_img_data.append(image.copy())
    pass
pass

flood_img_data = np.array(flood_img_data)

flood_target = pd.DataFrame({'target' :  np.repeat((1), len(flood_img_data))})
print(flood_target)

# calling the permanent water images

# assigning the path and list of age categories
# set the path of the folder where the images stored
# perm_data_HL_dir = "/content/drive/MyDrive/Sen1Flood11_New_Dataset/perm_train/0"
perm_data_HL_dir = "/content/drive/MyDrive/Sen1Flood11_New_Dataset/train/0/"


# set the list of categories available in the folder
perm_img_data = []

# iterate over each images, considering all the age groups
for img in os.listdir(perm_data_HL_dir):
  # read one by one image
    # print(img)
    try:
      img_array = keras.utils.load_img(os.path.join(perm_data_HL_dir, img), target_size=(32, 32))
    except:
      pass

    image = img_to_array(img_array)
    # # reshape data for the model
    image = image.reshape(( image.shape[0], image.shape[1], image.shape[2]))
    image = image.astype('float32')
    image /= 255
    # image = preprocess_input(image)
    perm_img_data.append(image.copy())
    # pass
pass

perm_img_data = np.array(perm_img_data)

perm_target = pd.DataFrame({'target' :  np.repeat((0), len(perm_img_data))})
print(perm_target)
```

```python
from keras.models import Sequential
from keras.layers import Activation, Dense, Flatten

model1 = Sequential()
model1.add(Flatten()) # first layer
# model.add(Conv2D(32,(3,3),activation='relu',input_shape=(1480, 32, 32, 3)))

# model.add(Flatten(input_shape=(32,32))) # first layer
model1.add(Dense(16))
model1.add(Activation('sigmoid'))
model1.add(Dense(12))
model1.add(Activation('relu'))
model1.add(Dense(10))
model1.add(Activation('softmax'))
model1.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
# model1.fit(x_train, y_train, epochs=200)
model1.fit(x_train, y_train, epochs=500,batch_size=128,validation_data=(x_valid,y_valid))
```
